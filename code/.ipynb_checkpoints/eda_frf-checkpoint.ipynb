```python
#let's define a function that I will be using to plot distribution of the other variables
def plot_distrib(series, title):
    plt.figure(figsize=(10,10))
    plt.hist(series, color='skyblue', bins=10)
    plt.axvline(series.median(), color = 'orange', lw=2)
    plt.xlabel(f'{title}', fontdict= {'fontsize':14})
    plt.ylabel('Frequency', fontdict= {'fontsize':14})
    plt.title(f'Distribution Of {title} In 10 Most Productive County, California', fontdict = {'fontsize':15})
    plt.tight_layout()
```

## EDA - Wine Production/ Climate Info  & Time Series

**Step1** On All California

```python
prod = df.copy()
prod = prod[['year','county', 'harvested acres', 'yield', 'production']]
prod = prod.groupby('year').sum().reset_index()
prod
```

```python
#let's turn date into datetine and index 
prod['year'] = prod['year'].map(lambda x: str(x) + '-' + '01' )
prod['year'] = pd.to_datetime(prod['year'] )
```

```python
prod.set_index('year',inplace=True)
```

```python
#let's define a function that I will be using to plot distribution of the other variables
def plot_time(series, title, yaxis):
    plt.figure(figsize=(10,6))
    plt.plot(series, color='skyblue')
    plt.xlabel('Time', fontdict= {'fontsize':14})
    plt.ylabel(f'{yaxis}', fontdict= {'fontsize':14})
    plt.title(f'{title} In 10 Most Productive County, California', fontdict = {'fontsize':15})
    plt.tight_layout()
plot_time(prod['production'], 'Wine Production', 'Production(tones)')
```

```python
plot_time(prod['harvested acres'], 'Wine Harvested', 'Harvest (Acres)')
```

```python
plot_time(prod['yield'], 'Wine Yield', 'Yield (tonnes/Acres)')
```

```python
#because I am interested in variability I want to standardize these  values.. let's do it on yields 
def stand_y(value):
    mean = np.mean(prod['yield'])
    std = np.std(prod['yield'])
    return ((value - mean)/std)
prod['yield_std'] = prod['yield'].map(stand_y)
```

```python
#because I am interested in variability I want to standardize these  values.. let's do it on production
def stand_p(value):
    mean = np.mean(prod['production'])
    std = np.std(prod['production'])
    return ((value - mean)/std)
prod['prod_std'] = prod['production'].map(stand_p)
```

```python
plot_time(prod['prod_std'], 'Standardized Wine Production', 'Std Production (unitless) ')
```

```python
plot_time(prod['yield_std'], 'Standardized Wine Yield', 'Std Yiled (unitless)')
```

```python
#and I want to compare these both to best potential predictors season var and annual var
climate = df[['year', 'season_var', 'annual_var']].iloc[:-10,:].copy()
```

```python
#let's turn date into datetine and index 
climate['year'] = climate['year'].map(lambda x: str(x) + '-' + '01' )
climate['year'] = pd.to_datetime(climate['year'])
```

```python
climate.set_index('year', inplace=True)
```

```python
def stand_s(value):
    mean = np.mean(climate['season_var'])
    std = np.std(climate['season_var'])
    return ((value - mean)/std)
def stand_a(value):
    mean = np.mean(climate['annual_var'])
    std = np.std(climate['annual_var'])
    return ((value - mean)/std)
climate['annual_var_std'] = climate['annual_var'].map(stand_a)
climate['season_var_std'] = climate['season_var'].map(stand_s)
```

```python
plot_time(climate['annual_var_std'], 'Annual Var', 'Nomralized Data(unitless)')
```

```python
plot_time(climate['season_var_std'], 'Season Var', 'Nomralized Data(unitless)')
```

```python
#Let's look at yield varibaility for the three most econmic import county  Napa, Sonoma, San Joaquin
#lets create single db
```

```python
sj = df2[df2['county']=='San Joaquin']
napa = df2[df2['county']=='Napa']
sonoma = df2[df2['county']=='Sonoma']
```

```python
#let's define a function that I will be using to plot distribution of the other variables
def plot_time_county(series1, series2, series3, title, yaxis, county):
    plt.figure(figsize=(10,6))
    plt.plot(series1, color='skyblue', label=f'Napa')
    plt.plot(series2, color='blue', label=f'Sonoma')
    plt.plot(series3, color='green', label=f'San Joaquin')
    plt.xlabel('Time', fontdict= {'fontsize':14})
    plt.ylabel(f'{yaxis}', fontdict= {'fontsize':14})
    plt.title(f'{title} In {county}, California', fontdict = {'fontsize':15})
    plt.legend()
    plt.tight_layout()
```

```python
plot_time_county(napa['production'], sonoma['production'],sj['production'], 
                 'Wine Production', 'Production(tones)', 'San Joaquin, Napa, Sonoma')
```

```python
plot_time_county(napa['yield'], sonoma['yield'],sj['yield'], 
                 'Wine Production', 'Yield', 'San Joaquin, Napa, Sonoma')
```

```python
plot_time_county(napa['value'], sonoma['value'],sj['value'], 
                 'Value', 'Value', 'San Joaquin, Napa, Sonoma')
```

```python
plot_time_county(napa['annual_tavg'], sonoma['annual_tavg'],sj['annual_tavg'], 
                 'Annual Temperature Average', 'Temperature Average', 'San Joaquin, Napa, Sonoma')
```

```python
plot_time_county(napa['annual_tmax'], sonoma['annual_tmax'],sj['annual_tmax'], 
                 'Annual Temperature Max', 'Temperature Max', 'San Joaquin, Napa, Sonoma')
```

```python
plot_time_county(napa['annual_precip'], sonoma['annual_precip'],sj['annual_precip'], 
                 'Annual Precipitation', 'Annual Precipitation', 'San Joaquin, Napa, Sonoma')
```
