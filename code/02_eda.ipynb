```python
def plot_per_county(column):
    for county in df['county'].unique():
        focus=df[df['county']==county]
        plt.plot(focus['year'],focus[column],label=county)
    plt.title(f'{column} per Year')
    plt.legend();
```

```python
def per_metric(metric):
    x=[]
    y=[]
    for year in range(2000,2020):
        x.append(year)
        focus=df[df['year']==year]
        y.append(focus[metric].mean())
    plt.plot(x,y,label=metric)
    
```

```python
per_metric('annual_precip')
per_metric('annual_tavg')
per_metric('annual_tmin')
per_metric('annual_tmax')
per_metric('annual_var')
per_metric('yield')

plt.title('Annual Metrics per Year')
plt.legend();
```
